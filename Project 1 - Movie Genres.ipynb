```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
imdb = pd.read_csv('imdb_movies.csv')
imdb.head()
```

```python
imdb.info()
```

```python
imdb['release_date'] = pd.to_datetime(imdb['release_date'], format = '%m/%d/%y')
imdb.info()
```

```python
imdb[['budget','revenue']] = imdb[['budget','revenue']].astype(dtype = 'float')
imdb.info()
```

```python
imdb.head()
```

```python
imdb.head(n = 3)
imdb['genres'].str.split('|')
```

Chech for duplicates


```python
imdb[imdb.duplicated()]
imdb.drop_duplicates(inplace= True)
```

```python
imdb[imdb.duplicated()]
```

```python
imdb.dropna(subset = 'genres', inplace= True)
```

```python
imdb.info()
```

```python
imdb.head(n = 3)
```

```python
imdb['profit'] = imdb['revenue'] - imdb['budget']
```

```python
imdb.head(n = 3)
```

```python
imdb_new_df = imdb[['popularity', 'budget','revenue', 'original_title','runtime','genres','release_date', 'vote_count','vote_average', 'profit']]
imdb_new_df.head()
```

```python
from pandas import Series, DataFrame
```

```python
split = imdb_new_df['genres'].str.split('|').apply(Series, 1).stack()
split
```

```python
split.name = 'genres_split'
del imdb_new_df['genres']
imdb_new_df = imdb_new_df.join(split)
```

```python
imdb_new_df
imdb_new_df['release_date'] = pd.to_datetime(imdb_new_df['release_date'], format = '%m/%d/%y')
imdb_new_df.head()
```

# 6. The best movies according to vote avg. return high profit and revenue.

```python
imdb_new_df = imdb[['popularity', 'budget','revenue', 'original_title','runtime','genres','release_date', 'vote_count','vote_average', 'profit']]
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
imdb_new_df.head()
```

```python
# We need to filter down to include movies that at least have 50 or 100 votes in order to avoid noise from movies with low number of ratings
movies_counted = imdb_new_df[imdb_new_df['vote_count'] >= 50]
movies_counted
```

```python
# Looking into correlation
movies_counted.corr(method = 'spearman', numeric_only = True)
```

```python
sns.regplot(data = movies_counted ,x = 'vote_average', y = 'profit', line_kws= {'color': 'green'})
plt.title('Spearman Correlation Between Vote AVG & Profit')
plt.show()
```

```python
sns.regplot(data = movies_counted ,x = 'vote_average', y = 'revenue', line_kws= {'color': 'orange'})
plt.title('Spearman Correlation Between Vote AVG & Revenue')
plt.show()
```

# 10. Porifts by Genre per Year

```python
imdb_new_df = imdb[['popularity', 'budget','revenue', 'original_title','runtime','genres','release_year', 'vote_count','vote_average', 'profit']]
imdb_new_df.head()
```

```python
split = imdb_new_df['genres'].str.split('|').apply(Series, 1).stack()
split
```

```python
split.index = split.index.droplevel(-1)
split
```

```python
split.name = 'genres_split'
del imdb_new_df['genres']
imdb_new_df = imdb_new_df.join(split)
imdb_new_df
```

```python
time_genre = pd.DataFrame(imdb_new_df.groupby(['release_year','genres_split'])['profit'].sum())
time_genre
```

```python
final_genre = pd.pivot_table(time_genre, values='profit', index = ['genres_split'], columns = ['release_year'])
final_genre
```

```python
sns.set(rc = {'figure.figsize': (10,15)})
sns.heatmap(data = final_genre, cmap= 'coolwarm', linewidths = 0.5)
plt.title('Genres by Profit per Year')
plt.show()
```

```python
sns.set(rc = {'figure.figsize': (10,15)})
sns.heatmap(data = final_genre, cmap= 'YlGnBu', linewidths = 0.5)
plt.title('Genres by Profit per Year')
plt.show()
```
